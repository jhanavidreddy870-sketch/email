```python
import pandas as pd

data ={
    "email":[
        "win money now",
        "hello how are you",
    "clain your free prize",
       "meeting at spm",
        "congratulation you won the lettery",
        "let's catch up today",
        "free enter in contest",
        "project discussion today"
        ],
    "label": [1,0,1,0,1,0,1,0]

}

df = pd.DataFrame(data)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df["email"])
y = df["label"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

x_train, x_test , y_train, y_test = train_test_split(X,y,test_size=0.2)

dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier()
gb.fit(x_train,y_train)
```

```python
test_email = ["meeting at spm"]
test_vector = vectorizer.transform(test_email)

def explain(pred):
  return "spam" if pred == 1 else "not spam"
print("DT:",explain(dt.predict(test_vector)[0]))
print("RF:",explain(pred=rf.predict(test_vector)[0]))
print("GB:",explain(pred=gb.predict(test_vector)[0]))
```
